# Problem

$$f(w_1, w_2) = 0.1w_1^2 + 2w_2^2 \;\;\;\;\;\;\;(1)$$

```python
import numpy as np
```

```python
def df_w(W):
    w1 = 0.2 * W[0]
    w2 = 4 * W[1]
    dW = np.array([w1, w2])
    return dW
```

```python
def sgd(W, dW, lr):
    W = W - lr * dW
    return W
```

```python
def train_p1(optimizer, lr, epochs):
    W = np.array([-5, -2], dtype=np.float32)
    results = [W]
    for i in range(epochs):
        dW = df_w(W)
        W = optimizer(W, dW, lr)
        results.append(W)
    return results
```

```python
train_p1(sgd, lr=0.4, epochs=30)
```
